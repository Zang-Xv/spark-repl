关于模型下载的各种问题和单元格

## WebVectors

搜索相关词汇的词云大模型[WebVectors](http://vectors.nlpl.eu/explore/embeddings/en/models/)

通过预训练的词汇向量集合来判断词汇的相似度

[English Wikipedia模型下载链接](https://vectors.nlpl.eu/repository/20/223.zip)

## DIS

移除图像背景的模型[DIS](https://github.com/xuebinqin/DIS)

处理图像, 最终结果是前景图像(智能主题抠图)

[谷歌云下载链接](https://drive.google.com/file/d/1O1eIuXX1hlGsV7qx4eSkjH231q7G1by1/view)
